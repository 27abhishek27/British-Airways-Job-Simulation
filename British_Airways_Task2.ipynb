### TASK #3: Mutual Information

```python

X= customer_booking_df.drop('booking_complete',axis=1)
y= customer_booking_df.booking_complete         

#changing object dtype to int dtype
for colname in X.select_dtypes("object"):
    X[colname], _ = X[colname].factorize()
```

```python
X.dtypes
```

```python
from sklearn.feature_selection import mutual_info_classif

mi_scores = mutual_info_classif(X, y)
mi_scores = pd.Series(mi_scores, name="MI Scores", index=X.columns)
mi_scores = mi_scores.sort_values(ascending=False)

mi_scores # show a few features with their MI scores
```

```python
def plot_mi_scores(scores):
    scores = scores.sort_values(ascending=True)
    width = np.arange(len(scores))
    ticks = list(scores.index)
    plt.barh(width, scores)
    plt.yticks(width, ticks)
    plt.title("Mutual Information Scores")


plt.figure(dpi=100, figsize=(8, 5))
plot_mi_scores(mi_scores)
```

### TASK #4: CREATE TESTING AND TRAINING DATASET & PERFORM DATA CLEANING

```python
customer_booking_df
```

```python
# Checking all categorical data
X_cat = customer_booking_df[['sales_channel','trip_type','flight_day','route','booking_origin']]
```

```python
X_cat
# we need to change this categorical data to numbers
```

```python
# importing important library to change categorical data to number
from sklearn.preprocessing import OneHotEncoder

onehotencoder = OneHotEncoder()

X_cat = onehotencoder.fit_transform(X_cat).toarray()
```

```python
# checking the changes in data
X_cat.shape
```

```python
# converting this new data to a dataframe
X_cat = pd.DataFrame(X_cat)
```

```python
X_cat
```

```python
# note that we dropped the target 'booking_complete'
X_numerical = customer_booking_df[['num_passengers', 'purchase_lead', 'length_of_stay', 'flight_hour', 'wants_extra_baggage', 'wants_preferred_seat', 'wants_in_flight_meals', 'flight_duration']]
```

```python
# doing the Concatenation the all categorical and numerical dataframe together
X_all = pd.concat([X_cat,X_numerical],axis=1)
```

```python
X_all
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

# X_all is a DataFrame, convert its column names to strings before applying MinMaxScaker
X_all.columns = X_all.columns.astype(str)

X = scaler.fit_transform(X_all)
```

```python
X
```

```python
# now we are putting our traget column in y variable from our customer_booking dataset
y = customer_booking_df['booking_complete']
y
```

### TASK #5: TRAIN AND EVALUATE A LOGISTIC REGRESSION CLASSIFIER

```python
from sklearn.model_selection import train_test_split

X_tarin, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25, random_state=101)
```

```python
# checking the tarin test split shape
X_tarin.shape
```

```python
X_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, roc_auc_score
```

```python
model = LogisticRegression()
model.fit(X_tarin,y_train)
```

```python
y_predictions = model.predict(X_test)
```

```python
# Testing Set Performance
y_predictions
```

```python
from sklearn.metrics import confusion_matrix, classification_report

# we have used .2f for only 2 decimal places
print(f'Accuracy {100 * accuracy_score(y_predictions, y_test):.2f} %')
print(f'AUC score {roc_auc_score(y_predictions, y_test):.2f} %')
```

```python
cm = confusion_matrix(y_predictions,y_test)
sns.heatmap(cm,annot=True)
```

```python
print(classification_report(y_test,y_predictions))
```

### TASK #6: TRAIN AND EVALUATE A RANDOM FOREST CLASSIFIER

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(X_tarin,y_train)
```

```python
y_predictions = model.predict(X_test)
```

```python
# Testing Set Performance
cm = confusion_matrix(y_test,y_predictions)
sns.heatmap(cm,annot=True)
```

```python
print(classification_report(y_predictions,y_test))
```
