### TASK #3: HANDLING MISSING VALUES

```python
# Fill up the missing elements with mean values of each columns
british_airways_df['Rating'] = british_airways_df['Rating'].fillna(british_airways_df['Rating'].mean())
british_airways_df['Seat Comfort'] = british_airways_df['Seat Comfort'].fillna(british_airways_df['Seat Comfort'].mean())
british_airways_df['Cabin Staff Service'] = british_airways_df['Cabin Staff Service'].fillna(british_airways_df['Cabin Staff Service'].mean())
british_airways_df['Food & Beverages'] = british_airways_df['Food & Beverages'].fillna(british_airways_df['Food & Beverages'].mean())
british_airways_df['Inflight Entertainment'] = british_airways_df['Inflight Entertainment'].fillna(british_airways_df['Inflight Entertainment'].mean())
british_airways_df['Ground Service'] = british_airways_df['Ground Service'].fillna(british_airways_df['Ground Service'].mean())
british_airways_df['Wifi & Connectivity'] = british_airways_df['Wifi & Connectivity'].fillna(british_airways_df['Wifi & Connectivity'].mean())
```

```python
# dropping the missing values in Location and Seat Type
british_airways_df = british_airways_df.dropna(subset=['Location', 'Seat Type'])
```

```python
# dropping the cloumns that are not usefull
british_airways_df = british_airways_df.drop(columns=['Name','Date','Aircraft', 'Type Of Traveller', 'Route', 'Date Flown'], axis=1)
```

```python
# now we don't have any missing values
british_airways_df.isnull().sum()
```

```python
# remove the ✅ 
british_airways_df['Verification'] = british_airways_df['Verification'].str.replace('✅', '', regex=False).str.strip()
```

```python
# Let's replace 'Verification' , 'Recommended' column with integers before performing any visualizations
british_airways_df['Verification'] = british_airways_df['Verification'].apply(lambda x:1 if x == 'Trip Verified' else 0)
british_airways_df['Recommended'] = british_airways_df['Recommended'].apply(lambda x:1 if x == 'yes' else 0)
```

```python
british_airways_df = pd.concat([british_airways_df, catergorial_data], axis=1)
```

```python
british_airways_df.head()
```

```python
import string
string.punctuation
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
# Let's define a pipeline to clean up all the messages
# The pipeline performs the following: (1) remove punctuation, (2) remove stopwords

def message_cleaning(message):
    Test_punc_removed = [ char for char in message if char not in string.punctuation]
    Test_punc_removed_join = ''.join(Test_punc_removed)
    Test_punc_removed_join_clean = [word for word in Test_punc_removed_join.split() if word.lower() not in stopwords.words('english')]
    return Test_punc_removed_join_clean
```

```python
# Let's test the newly added function
british_airways_df_clean = british_airways_df['Full Review'].apply(message_cleaning)
```

```python
print(british_airways_df_clean[3])
```

```python
british_airways_df_clean
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(analyzer = message_cleaning)
full_review_countvectorizer = vectorizer.fit_transform(british_airways_df['Full Review'])
```

### TASK #6: TRAIN A NAIVE BAYES CLASSIFIER MODEL

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
NB_classifier = MultinomialNB()

X_train.columns = X_train.columns.astype(str)
X_test.columns = X_test.columns.astype(str)

NB_classifier.fit(X_train, y_train)
```

### TASK #7: ASSESS TRAINED MODEL PERFORMANCE

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
y_predict_test = NB_classifier.predict(X_train)
y_predict_test
```

```python
cm = confusion_matrix(y_train, y_predict_test)
sns.heatmap(cm, annot=True)
```

```python
y_predict_test = NB_classifier.predict(X_test)
cm = confusion_matrix(y_test, y_predict_test)
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_test, y_predict_test))
```
